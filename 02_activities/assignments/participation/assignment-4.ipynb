9. What ‘underwater labour’ contributed to your final data visualization product?

The ‘underwater labour’ included cleaning the dataset to handle encoding issues, removing missing values, and aggregating data by city. These preprocessing steps were essential for creating an accurate and meaningful bar chart.

This bar chart effectively visualizes the distribution of settlement organizations across cities, providing valuable insights for decision-making and planning in the context of settlement services.


```python
import pandas as pd

# Load the dataset with different encoding options
try:
    data = pd.read_csv('../../data/settlement_services_may2024_en.csv', encoding='latin1')
except UnicodeDecodeError:
    import chardet
    with open('../../data/settlement_services_may2024_en.csv', 'rb') as f:
        result = chardet.detect(f.read())
    data = pd.read_csv('../../data/settlement_services_may2024_en.csv', encoding=result['encoding'])
    
# Display the first few rows
print(data.head())

# Clean any leading/trailing spaces in column names
data.columns = data.columns.str.strip()

# Handle missing values if needed
data = data.dropna()  # Drop rows with missing values

# Reset index after dropping rows
data.reset_index(drop=True, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Count the number of organizations by city
city_counts = data['City'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 8))
city_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Settlement Organizations by City')
plt.xlabel('City')
plt.ylabel('Number of Organizations')
plt.xticks(rotation=90)
plt.tight_layout()

# Save the plot
plt.savefig('settlement_organizations_by_city.png')

# Show the plot
plt.show()
```

# Visualization 2: Pie Chart of Settlement Services Across Cities in Ontario

1. What software did you use to create your data visualization?

I used Plotly in Python to create the pie chart. Plotly is a powerful and interactive graphing library that allows for detailed and aesthetically pleasing visualizations.

2. Who is your intended audience?

The intended audience includes policymakers, local government officials, and community organizations in Ontario. These stakeholders are interested in understanding the distribution of settlement services across various cities to better allocate resources and support community needs.

3. What information or message are you trying to convey with your visualization?

The pie chart aims to convey the distribution of settlement services among different cities in Ontario. It provides a visual representation of how services are spread out geographically, highlighting which cities have the highest and lowest number of services. This helps in identifying areas with potentially higher or lower service availability.

4. What design principles (substantive, perceptual, aesthetic) did you consider when making your visualization? How did you apply these principles? With what elements of your plots?

Substantive Principles: The pie chart accurately reflects the number of settlement services by city. I applied this principle by ensuring the chart displays clear and accurate data, including aggregating smaller categories into an “Other” category for clarity.

Perceptual Principles: The chart uses contrasting colors for different cities, enhancing visual differentiation and making it easy to interpret. The use of a donut style (with a hole) improves readability by focusing attention on the proportions rather than the absolute values.

Aesthetic Principles: The design includes a clean layout with a central title, color-coded segments, and a legend placed strategically to avoid clutter. This approach enhances visual appeal and makes the chart easy to understand.

5. How did you ensure that your data visualizations are reproducible? If the tool you used to make your data visualization is not reproducible, how will this impact your data visualization?

Plotly charts created in Python are reproducible by sharing the script or code that generates them. This ensures that others can recreate the exact same visualization by running the code with the same dataset. If the visualization tool were not reproducible, it could lead to inconsistencies and difficulties in verifying or updating the visualization, impacting its reliability.

6. How did you ensure that your data visualization is accessible?

The pie chart is designed to be accessible by using distinct colors and a clear legend. The choice of colors ensures that people with color vision deficiencies can still differentiate between segments. Additionally, the chart includes labels and a legend to provide context for each segment.

7. Who are the individuals and communities who might be impacted by your visualization?

Individuals and communities impacted include residents of Ontario, especially those in cities with fewer services, who might benefit from targeted resource allocation. Local government officials and community organizations can use the information to plan and enhance the delivery of settlement services.

8. How did you choose which features of your chosen dataset to include or exclude from your visualization?

I included features related to the city and the number of settlement services to focus on the distribution and availability of services. Features such as specific addresses or websites were excluded to simplify the visualization and concentrate on the overall distribution by city.

9. What ‘underwater labour’ contributed to your final data visualization product?

The ‘underwater labour’ involved cleaning and preprocessing the dataset to ensure accuracy. This included handling encoding issues, dealing with missing values, and aggregating smaller cities into an “Other” category for clarity. These steps were crucial for producing a clear and informative pie chart.

This pie chart visualization effectively communicates the distribution of settlement services across cities in Ontario, providing valuable insights for decision-making and resource allocation.

```python
import pandas as pd
import plotly.express as px


# Load and clean data
data = pd.read_csv('../../data/settlement_services_may2024_en.csv', encoding='ISO-8859-1')
data.columns = data.columns.str.strip()
data_on = data[data['Province'] == 'ON']
data_on = data_on.dropna(subset=['City'])

# Count the number of services per city
city_counts = data_on.groupby('City').size().reset_index(name='Number of Services')

# If there are too many cities, group smaller ones into 'Other'
threshold = 5  # Minimum number of services to display individually
city_counts = city_counts.sort_values('Number of Services', ascending=False)
city_counts['City'] = city_counts['City'].where(city_counts['Number of Services'] > threshold, 'Other')
city_counts = city_counts.groupby('City').sum().reset_index()

# Plot the pie chart
fig = px.pie(
    city_counts, 
    names='City', 
    values='Number of Services', 
    title='Distribution of Settlement Services Across Cities in Ontario',
    color='City',
    color_discrete_sequence=px.colors.qualitative.Plotly,
    labels={'Number of Services': 'Number of Services'},
    hole=0.3  # Adding a hole makes it a donut chart for better visibility
)

# Update layout for better visualization
fig.update_layout(
    title_font_size=20,
    title_x=0.5,
    legend_title='City',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

# Display the plot
fig.show()

# Please see img ../participation/vizualisation_2_result_pie_chart.png
```
