```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# Activity 

# Create a line plot with both y variables
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y1)
ax.plot(x, y2)
ax.legend(loc='lower right')  # Adding a legend

# Show the plot
plt.show()
```

```python
# Adding a legend 

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
# Modifying a legend

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right'
,
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True )
fig.show()
```

```python
# Moving the legend outside of the plot area

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc= 'upper left',
bbox_to_anchor =(1, 1))
fig.show()
```

```python
# Annotations with the text() function

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95,
"This value is important!")
fig.show()
```

```python
# Modifying text annotations

ax.text(10, 95,
"This value is important!",
ha='center', ##alignment of text
color = 'red', ##modify font colour
size = 20) ##modify font size

fig.show()
```

```python
# Positioning text annotations

# Start by making a simple, empty set of axes

fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5,". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1,". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2,". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# Annotating with arrows

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black', shrink=0.05))

fig.show()
```

```python
# Modifying arrows

ax.annotate('This is important!',
            xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(arrowstyle="wedge",
                            color="hotpink"))

fig.show()
```

```python
# Removing tick marks and labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Limiting tick marks

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Tick mark intervals

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotating axis labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
plt.xticks(rotation=45, ha='right')

# Rotate x-axis labels and modify their alignment and font size
plt.xticks(rotation=45, ha='right', fontsize=12, fontname='serif')

plt.show()
```

```python
# Activity

import numpy as np
import matplotlib.pyplot as plt

# Set up sample data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(10, 5))

# Customize scatter plot
ax1.scatter(x1, y1, color='purple', marker='^', edgecolor='black', s=100)
ax1.set_title('Scatter Plot', fontsize=14, fontweight='bold')
ax1.set_xlabel('X Axis', fontsize=12, fontdict={'family': 'serif', 'color': 'indigo'})
ax1.set_ylabel('Y Axis', fontsize=12, fontdict={'family': 'serif', 'color': 'indigo'})
ax1.text(10, 70, 'High Value', color='red', fontsize=10, ha='center')
ax1.grid(True)

# Customize bar plot
ax2.bar(x2, y2, color=['red', 'blue', 'green', 'orange', 'purple'])
ax2.set_title('Bar Plot', fontsize=14, fontweight='bold')
ax2.set_xlabel('Characters', fontsize=12, fontdict={'family': 'serif', 'color': 'indigo'})
ax2.set_ylabel('Scores', fontsize=12, fontdict={'family': 'serif', 'color': 'indigo'})
ax2.annotate('Highest', xy=(2, 240), xytext=(3, 260),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='blue', ha='center')
ax2.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1'
,
'ax3'],
['ax2'
,
'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], ['ax2','ax3']], figsize=(7, 4), layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label'
, fontsize=18)
someaxes["ax2"].set_xlabel('Another Label'
, fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious'
, fontsize=18)
plt.show()
```

```python
# Images

from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
                          0.11, # y coordinate (ON FIGURE, NOT AXES)
                          0.15, # image width
                          0.35]) # image height


fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "red")
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])
ax_image.imshow(image)
ax_image.axis("off")
plt.show()
```
